```python
import numpy as np
import matplotlib.pyplot as plt

# Função para simular jogadas de dados
def simular_jogadas(num_jogadas, num_faces):
    return np.random.randint(1, num_faces + 1, num_jogadas)

# Simulando jogadas
num_jogadas = 10000
jogadas_d12 = simular_jogadas(num_jogadas, 12)
jogadas_2d6 = simular_jogadas(num_jogadas, 6) + simular_jogadas(num_jogadas, 6)

# Calculando o valor esperado
valor_esperado_d12 = np.mean(jogadas_d12)
valor_esperado_2d6 = np.mean(jogadas_2d6)

print(f"Valor esperado para 1 dado de 12 faces: {valor_esperado_d12:.2f}")
print(f"Valor esperado para 2 dados de 6 faces: {valor_esperado_2d6:.2f}")

# Gráfico de barras para 1 dado de 12 faces
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(jogadas_d12, bins=np.arange(1, 14)-0.5, rwidth=0.8, color='skyblue')
plt.title('Distribuição - 1 Dado de 12 Faces')
plt.xlabel('Valor do Dado')
plt.ylabel('Frequência')
plt.xticks(np.arange(1, 13))

# Gráfico de barras para 2 dados de 6 faces
plt.subplot(1, 2, 2)
plt.hist(jogadas_2d6, bins=np.arange(2, 14)-0.5, rwidth=0.8, color='salmon')
plt.title('Distribuição - 2 Dados de 6 Faces')
plt.xlabel('Valor do Dado')
plt.ylabel('Frequência')
plt.xticks(np.arange(2, 13))

plt.tight_layout()
plt.show()
```

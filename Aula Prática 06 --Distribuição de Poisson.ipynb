```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.stats import binom
import seaborn as sns
from scipy.stats import norm
```

```python
# Questão 1

# Parâmetro lambda (taxa média de ocorrências)
lambda_value = 2

# (a) Probabilidade de exatamente 3 pessoas sofrerem a reação
#     poisson.pmf(k, λ) = (e^(-λ) * λ^k) / k!
prob_exatamente_3 = poisson.pmf(3, lambda_value) 

# (b) Probabilidade de mais do que 2 pessoas sofrerem a reação
prob_mais_que_2 = 1 - poisson.cdf(2, lambda_value)

print(f"Probabilidade de exatamente 3 pessoas sofrerem a reação: {prob_exatamente_3:.4f}")
print(f"Probabilidade de mais do que 2 pessoas sofrerem a reação: {prob_mais_que_2:.4f}")
```

```python
# Questão 2

# Parâmetro lambda (média de acidentes)
lambda_value = 6

# (a) Probabilidade de 0 acidentes
prob_0_acidentes = poisson.pmf(0, lambda_value)

# (b) Probabilidade de 2 acidentes
prob_2_acidentes = poisson.pmf(2, lambda_value)

# (c) Probabilidade de 6 acidentes
prob_6_acidentes = poisson.pmf(6, lambda_value)

# (d) Probabilidade de 8 acidentes
prob_8_acidentes = poisson.pmf(8, lambda_value)

# (e) Probabilidade de 10 acidentes
prob_10_acidentes = poisson.pmf(10, lambda_value)

# (f) Probabilidade de entre 4 e 10 acidentes
prob_4_a_10 = poisson.cdf(10, lambda_value) - poisson.cdf(3, lambda_value)

# Exibir resultados
print(f"Probabilidade de 0 acidentes: {prob_0_acidentes:.4f}")
print(f"Probabilidade de 2 acidentes: {prob_2_acidentes:.4f}")
print(f"Probabilidade de 6 acidentes: {prob_6_acidentes:.4f}")
print(f"Probabilidade de 8 acidentes: {prob_8_acidentes:.4f}")
print(f"Probabilidade de 10 acidentes: {prob_10_acidentes:.4f}")
print(f"Probabilidade de entre 4 e 10 acidentes: {prob_4_a_10:.4f}")

# Plotando o histograma
k_values = np.arange(0, 11)
pmf_values = poisson.pmf(k_values, lambda_value)
```

```python
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.xlabel('Número de Acidentes')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Poisson: Número de Acidentes na Rodovia')
plt.show()
```

```python
# Questão 3

# Parâmetro lambda (média esperada de peças defeituosas)
lambda_value = 3

# Cálculo das probabilidades para 0, 1, 2, 3 e 4 peças defeituosas usando Poisson
prob_0_defeituosas_poisson = poisson.pmf(0, lambda_value)
prob_1_defeituosa_poisson = poisson.pmf(1, lambda_value)
prob_2_defeituosas_poisson = poisson.pmf(2, lambda_value)
prob_3_defeituosas_poisson = poisson.pmf(3, lambda_value)
prob_4_defeituosas_poisson = poisson.pmf(4, lambda_value)

# Exibir resultados
print(f"Poisson - Probabilidade de 0 defeituosas: {prob_0_defeituosas_poisson:.4f}")
print(f"Poisson - Probabilidade de 1 defeituosa: {prob_1_defeituosa_poisson:.4f}")
print(f"Poisson - Probabilidade de 2 defeituosas: {prob_2_defeituosas_poisson:.4f}")
print(f"Poisson - Probabilidade de 3 defeituosas: {prob_3_defeituosas_poisson:.4f}")
print(f"Poisson - Probabilidade de 4 defeituosas: {prob_4_defeituosas_poisson:.4f}")
```

```python
# Questão 4

# Parâmetros da distribuição binomial
n = 10  # número de peças na amostra
p = 0.1  # probabilidade de defeito em uma peça
k = 2    # número de peças defeituosas

# Probabilidade de exatamente 2 defeituosas
prob_binomial = binom.pmf(k, n, p)
print(f"Probabilidade binomial de exatamente 2 defeituosas: {prob_binomial:.4f}")
```

```python
# Parâmetro lambda para a distribuição de Poisson
lambda_value = n * p  # média de defeitos esperados

# Probabilidade de exatamente 2 defeituosas
prob_poisson = poisson.pmf(k, lambda_value)
print(f"Probabilidade de Poisson de exatamente 2 defeituosas: {prob_poisson:.4f}")
```

## Porque é possível considerar estas duas distribuições. Qual a principal diferença entre essas distribuições e a distribuição normal?

#### A distribuição de Poisson se torna vantajosa em relação à distribuição binomial quando se observam as seguintes condições:

#### Número grande de tentativas (n):

A binomial requer um cálculo mais complexo conforme o número de tentativas n aumenta. Para valores muito grandes de n, como 1000, 10000 ou mais, a fórmula da binomial torna-se computacionalmente cara.

A distribuição de Poisson é mais simples e direta para esses casos, uma vez que o único parâmetro necessário é a média λ, independentemente de n.

#### Probabilidade de sucesso pequena (p):

Quando a probabilidade de sucesso p é muito pequena (tipicamente menor que 0,05), a binomial pode ser aproximada por Poisson.
O Poisson lida bem com eventos raros, como acidentes ou defeitos, que têm baixa probabilidade em uma grande população.
Média moderada de sucessos (𝜆=𝑛×𝑝):

A aproximação de Poisson funciona bem quando o número esperado de sucessos 𝜆 é um valor moderado. Isso significa que, mesmo com muitos eventos, os sucessos são poucos em comparação ao total.

```python
# Função para gerar os gráficos
def plot_poisson_normal_fit(lambda_value, size=10000):
    # Gerando números aleatórios seguindo a distribuição de Poisson
    data_poisson = np.random.poisson(lambda_value, size)

    # Criando o histograma da distribuição de Poisson
    sns.histplot(data_poisson, bins=30, kde=False, color='Skyblue', stat='density', label=f"Poisson λ={lambda_value}", alpha=0.6)

    # Ajuste de uma curva normal
    mu = np.mean(data_poisson)
    sigma = np.std(data_poisson)
    
    # Gerando valores da distribuição normal com a mesma média e desvio padrão
    x = np.linspace(min(data_poisson), max(data_poisson), 100)
    plt.plot(x, norm.pdf(x, mu, sigma), 'r-', label=f'Normal fit (μ={mu:.2f}, σ={sigma:.2f})', lw=2)
    
    # Configurações do gráfico
    plt.title(f'Distribuição de Poisson e Ajuste Normal para λ = {lambda_value}')
    plt.xlabel('Número de eventos')
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

# Gerar e plotar para diferentes valores de λ
for lambda_value in [1, 5, 10, 20, 50]:
    plot_poisson_normal_fit(lambda_value)
```

```python
# Função para gerar e plotar as distribuições
def plot_distributions(n, p, lambda_value, size=10000):
    # Gerando dados binomiais
    data_binomial = np.random.binomial(n=n, p=p, size=size)
    
    # Gerando dados Poisson
    data_poisson = np.random.poisson(lambda_value, size=size)
    
    # Gerando dados Normais
    data_normal = np.random.normal(loc=lambda_value, scale=np.sqrt(lambda_value), size=size)

    # Criar o histograma da distribuição binomial
    sns.histplot(data_binomial, bins=30, kde=False, color='blue', stat='density', label=f"Binomial n={n}, p={p}", alpha=0.6)
    
    # Criar o histograma da distribuição Poisson
    sns.histplot(data_poisson, bins=30, kde=False, color='green', stat='density', label=f"Poisson λ={lambda_value}", alpha=0.6)
    
    # Criar o histograma da distribuição Normal
    sns.histplot(data_normal, bins=30, kde=False, color='red', stat='density', label=f"Normal μ={lambda_value}, σ={np.sqrt(lambda_value):.2f}", alpha=0.6)

    # Configurações do gráfico
    plt.title(f'Comparação: Distribuições Binomial, Poisson e Normal')
    plt.xlabel('Número de eventos')
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

# Exemplo 1: Valores moderados (as distribuições começam a se aproximar)
plot_distributions(n=100, p=0.05, lambda_value=5)

# Exemplo 2: Valores maiores para λ (aproximação mais forte entre Poisson e Normal)
plot_distributions(n=500, p=0.01, lambda_value=5)
```

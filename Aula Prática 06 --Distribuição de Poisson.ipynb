```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.stats import binom
import seaborn as sns
from scipy.stats import norm
```

```python
# Quest√£o 1

# Par√¢metro lambda (taxa m√©dia de ocorr√™ncias)
lambda_value = 2

# (a) Probabilidade de exatamente 3 pessoas sofrerem a rea√ß√£o
#     poisson.pmf(k, Œª) = (e^(-Œª) * Œª^k) / k!
prob_exatamente_3 = poisson.pmf(3, lambda_value) 

# (b) Probabilidade de mais do que 2 pessoas sofrerem a rea√ß√£o
prob_mais_que_2 = 1 - poisson.cdf(2, lambda_value)

print(f"Probabilidade de exatamente 3 pessoas sofrerem a rea√ß√£o: {prob_exatamente_3:.4f}")
print(f"Probabilidade de mais do que 2 pessoas sofrerem a rea√ß√£o: {prob_mais_que_2:.4f}")
```

```python
# Quest√£o 2

# Par√¢metro lambda (m√©dia de acidentes)
lambda_value = 6

# (a) Probabilidade de 0 acidentes
prob_0_acidentes = poisson.pmf(0, lambda_value)

# (b) Probabilidade de 2 acidentes
prob_2_acidentes = poisson.pmf(2, lambda_value)

# (c) Probabilidade de 6 acidentes
prob_6_acidentes = poisson.pmf(6, lambda_value)

# (d) Probabilidade de 8 acidentes
prob_8_acidentes = poisson.pmf(8, lambda_value)

# (e) Probabilidade de 10 acidentes
prob_10_acidentes = poisson.pmf(10, lambda_value)

# (f) Probabilidade de entre 4 e 10 acidentes
prob_4_a_10 = poisson.cdf(10, lambda_value) - poisson.cdf(3, lambda_value)

# Exibir resultados
print(f"Probabilidade de 0 acidentes: {prob_0_acidentes:.4f}")
print(f"Probabilidade de 2 acidentes: {prob_2_acidentes:.4f}")
print(f"Probabilidade de 6 acidentes: {prob_6_acidentes:.4f}")
print(f"Probabilidade de 8 acidentes: {prob_8_acidentes:.4f}")
print(f"Probabilidade de 10 acidentes: {prob_10_acidentes:.4f}")
print(f"Probabilidade de entre 4 e 10 acidentes: {prob_4_a_10:.4f}")

# Plotando o histograma
k_values = np.arange(0, 11)
pmf_values = poisson.pmf(k_values, lambda_value)
```

```python
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.xlabel('N√∫mero de Acidentes')
plt.ylabel('Probabilidade')
plt.title('Distribui√ß√£o de Poisson: N√∫mero de Acidentes na Rodovia')
plt.show()
```

```python
# Quest√£o 3

# Par√¢metro lambda (m√©dia esperada de pe√ßas defeituosas)
lambda_value = 3

# C√°lculo das probabilidades para 0, 1, 2, 3 e 4 pe√ßas defeituosas usando Poisson
prob_0_defeituosas_poisson = poisson.pmf(0, lambda_value)
prob_1_defeituosa_poisson = poisson.pmf(1, lambda_value)
prob_2_defeituosas_poisson = poisson.pmf(2, lambda_value)
prob_3_defeituosas_poisson = poisson.pmf(3, lambda_value)
prob_4_defeituosas_poisson = poisson.pmf(4, lambda_value)

# Exibir resultados
print(f"Poisson - Probabilidade de 0 defeituosas: {prob_0_defeituosas_poisson:.4f}")
print(f"Poisson - Probabilidade de 1 defeituosa: {prob_1_defeituosa_poisson:.4f}")
print(f"Poisson - Probabilidade de 2 defeituosas: {prob_2_defeituosas_poisson:.4f}")
print(f"Poisson - Probabilidade de 3 defeituosas: {prob_3_defeituosas_poisson:.4f}")
print(f"Poisson - Probabilidade de 4 defeituosas: {prob_4_defeituosas_poisson:.4f}")
```

```python
# Quest√£o 4

# Par√¢metros da distribui√ß√£o binomial
n = 10  # n√∫mero de pe√ßas na amostra
p = 0.1  # probabilidade de defeito em uma pe√ßa
k = 2    # n√∫mero de pe√ßas defeituosas

# Probabilidade de exatamente 2 defeituosas
prob_binomial = binom.pmf(k, n, p)
print(f"Probabilidade binomial de exatamente 2 defeituosas: {prob_binomial:.4f}")
```

```python
# Par√¢metro lambda para a distribui√ß√£o de Poisson
lambda_value = n * p  # m√©dia de defeitos esperados

# Probabilidade de exatamente 2 defeituosas
prob_poisson = poisson.pmf(k, lambda_value)
print(f"Probabilidade de Poisson de exatamente 2 defeituosas: {prob_poisson:.4f}")
```

## Porque √© poss√≠vel considerar estas duas distribui√ß√µes. Qual a principal diferen√ßa entre essas distribui√ß√µes e a distribui√ß√£o normal?

#### A distribui√ß√£o de Poisson se torna vantajosa em rela√ß√£o √† distribui√ß√£o binomial quando se observam as seguintes condi√ß√µes:

#### N√∫mero grande de tentativas (n):

A binomial requer um c√°lculo mais complexo conforme o n√∫mero de tentativas n aumenta. Para valores muito grandes de n, como 1000, 10000 ou mais, a f√≥rmula da binomial torna-se computacionalmente cara.

A distribui√ß√£o de Poisson √© mais simples e direta para esses casos, uma vez que o √∫nico par√¢metro necess√°rio √© a m√©dia Œª, independentemente de n.

#### Probabilidade de sucesso pequena (p):

Quando a probabilidade de sucesso p √© muito pequena (tipicamente menor que 0,05), a binomial pode ser aproximada por Poisson.
O Poisson lida bem com eventos raros, como acidentes ou defeitos, que t√™m baixa probabilidade em uma grande popula√ß√£o.
M√©dia moderada de sucessos (ùúÜ=ùëõ√óùëù):

A aproxima√ß√£o de Poisson funciona bem quando o n√∫mero esperado de sucessos ùúÜ √© um valor moderado. Isso significa que, mesmo com muitos eventos, os sucessos s√£o poucos em compara√ß√£o ao total.

6) Gere um conjunto de dados aleat√≥rios distribu√≠dos de acordo com a distribui√ß√£o normal, binomial e de Poisson que lhe permitam concluir quando as 3 distribui√ß√µes se aproximam. e explique o por que de acontecer tal situa√ß√£o no final de forma resumida

```python
# Fun√ß√£o para gerar os gr√°ficos
def plot_poisson_normal_fit(lambda_value, size=10000):
    # Gerando n√∫meros aleat√≥rios seguindo a distribui√ß√£o de Poisson
    data_poisson = np.random.poisson(lambda_value, size)

    # Criando o histograma da distribui√ß√£o de Poisson
    sns.histplot(data_poisson, bins=30, kde=False, color='Skyblue', stat='density', label=f"Poisson Œª={lambda_value}", alpha=0.6)

    # Ajuste de uma curva normal
    mu = np.mean(data_poisson)
    sigma = np.std(data_poisson)
    
    # Gerando valores da distribui√ß√£o normal com a mesma m√©dia e desvio padr√£o
    x = np.linspace(min(data_poisson), max(data_poisson), 100)
    plt.plot(x, norm.pdf(x, mu, sigma), 'r-', label=f'Normal fit (Œº={mu:.2f}, œÉ={sigma:.2f})', lw=2)
    
    # Configura√ß√µes do gr√°fico
    plt.title(f'Distribui√ß√£o de Poisson e Ajuste Normal para Œª = {lambda_value}')
    plt.xlabel('N√∫mero de eventos')
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

# Gerar e plotar para diferentes valores de Œª
for lambda_value in [1, 5, 10, 20, 50]:
    plot_poisson_normal_fit(lambda_value)
```

```python
# Fun√ß√£o para gerar e plotar as distribui√ß√µes
def plot_distributions(n, p, lambda_value, size=10000):
    # Gerando dados binomiais
    data_binomial = np.random.binomial(n=n, p=p, size=size)
    
    # Gerando dados Poisson
    data_poisson = np.random.poisson(lambda_value, size=size)
    
    # Gerando dados Normais
    data_normal = np.random.normal(loc=lambda_value, scale=np.sqrt(lambda_value), size=size)

    # Criar o histograma da distribui√ß√£o binomial
    sns.histplot(data_binomial, bins=30, kde=False, color='blue', stat='density', label=f"Binomial n={n}, p={p}", alpha=0.6)
    
    # Criar o histograma da distribui√ß√£o Poisson
    sns.histplot(data_poisson, bins=30, kde=False, color='green', stat='density', label=f"Poisson Œª={lambda_value}", alpha=0.6)
    
    # Criar o histograma da distribui√ß√£o Normal
    sns.histplot(data_normal, bins=30, kde=False, color='red', stat='density', label=f"Normal Œº={lambda_value}, œÉ={np.sqrt(lambda_value):.2f}", alpha=0.6)

    # Configura√ß√µes do gr√°fico
    plt.title(f'Compara√ß√£o: Distribui√ß√µes Binomial, Poisson e Normal')
    plt.xlabel('N√∫mero de eventos')
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

# Exemplo 1: Valores moderados (as distribui√ß√µes come√ßam a se aproximar)
plot_distributions(n=100, p=0.05, lambda_value=5)

# Exemplo 2: Valores maiores para Œª (aproxima√ß√£o mais forte entre Poisson e Normal)
plot_distributions(n=500, p=0.01, lambda_value=5)
```

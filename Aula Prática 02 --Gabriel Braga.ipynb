```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ler o arquivo
tabela = pd.read_csv("TabelaSalarial.csv")
tabela
```

```python
# prencher os valores faltantes com a média em N de Filhos
tabela["N de Filhos"].fillna(tabela["N de Filhos"].mean(), inplace=True)
tabela
```

```python
# Retirar duplicadas
tabela.drop_duplicates(inplace=True)
tabela
```

```python
# Tirar colunas que não são necessárias para a análise com a "N"
tabela.drop(columns="N", inplace=True)
tabela
```

```python
# Questão 1

tabela.describe()
```

```python
# plotar histograma de todas as colunas "Estado Civil",	"Grau de Instrução", "N de Filhos",	"Salario (x Sal Min)", 
# "Idade (em anos)", "Região de Procedência" com seaborn dentro de um loop
colunas = ["Estado Civil",	"Grau de Instrução", "N de Filhos",	"Salario (x Sal Min)", "Idade (em anos)", "Região de Procedência"]
for coluna in colunas:
    sns.histplot(tabela[coluna])
    plt.title(coluna)
    plt.tight_layout()
    plt.show()
    
```

```python
# Faça o boxplot de todas as variáveis.
# separar as colunas em categoricas e numericas
categoricas = ["Estado Civil", "Grau de Instrução", "Região de Procedência"]
numericas = ["N de Filhos", "Salario (x Sal Min)", "Idade (em anos)"]

for coluna in categoricas:
    sns.boxplot(x=coluna, y="Salario (x Sal Min)", data=tabela)
    plt.title(coluna)
    plt.tight_layout()
    plt.show()
```

```python
for coluna in numericas:
    sns.boxplot(y=coluna, data=tabela)
    plt.title(coluna)
    plt.tight_layout()
    plt.show()
```
